```python
df.to_sql('sucursales', con=mi_conexion, index=False, index_label='id', if_exists='replace')
```

Id: Es el ID principal, compuesto por la concatenación de banderaId-comercioId-sucursalId.

```python
from sqlalchemy import text
with mi_conexion.connect() as connection:
    races = connection.execute(text("SELECT banderaDescripcion, banderaId FROM sucursales GROUP BY banderaId"))
    
    for row in races:
        print(row)
```

```python
with mi_conexion.connect() as connection:
    races = connection.execute(text("SELECT comercioId, comercioRazonSocial FROM sucursales GROUP BY comercioId"))
    
    for row in races:
        print(row)
```

```python
with mi_conexion.connect() as connection:
    races = connection.execute(text("SELECT  banderaId,banderaDescripcion FROM sucursales GROUP BY banderaId ORDER BY banderaId DESC"))
    
    for row in races:
        print(row)
```
